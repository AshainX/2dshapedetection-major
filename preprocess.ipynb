```python
''' 

CONTOUR IMAGE 

'''

import cv2
import os
import numpy as np

# Set the folder paths
input_folder = "/home/ashwathama/Documents/GitHub/2dshapedetection-major/testsegmentation"
output_folder = "/home/ashwathama/Documents/GitHub/2dshapedetection-major/testcontour"

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Function to preprocess and segment the image
def preprocess_and_segment_image(image):
    # Step 1: Apply Gaussian Blur to reduce noise
    blurred_img = cv2.GaussianBlur(image, (5, 5), 0)

    # Step 2: Morphological operation (Opening) to remove small noise
    kernel = np.ones((3, 3), np.uint8)  # Kernel for morphological operations
    morph_img = cv2.morphologyEx(blurred_img, cv2.MORPH_OPEN, kernel)

    # Step 3: Adaptive Thresholding for better segmentation
    threshold_img = cv2.adaptiveThreshold(morph_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY, 11, 2)

    # Step 4: Edge detection (Canny) to improve object boundaries
    edges_img = cv2.Canny(threshold_img, 50, 150)

    return edges_img

# Iterate through each file in the input folder
for filename in os.listdir(input_folder):
    file_path = os.path.join(input_folder, filename)

    # Ensure the file is an image (you can add more checks based on the image types)
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        # Read the binary image
        img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

        # Preprocess and segment the image
        segmented_img = preprocess_and_segment_image(img)

        # Find contours from the segmented edges
        contours, _ = cv2.findContours(segmented_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Create an empty image to draw contours
        contour_img = cv2.cvtColor(segmented_img, cv2.COLOR_GRAY2BGR)

        # Draw contours on the image
        cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 2)

        # Save the segmented image with contours drawn
        output_file_path = os.path.join(output_folder, f"segmented_{filename}")
        cv2.imwrite(output_file_path, contour_img)

        print(f"Processed {filename} and saved segmented image to {output_file_path}")

print("Image segmentation completed.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os


# Input and Output directories
input_dir = '/home/ashwathama/Documents/GitHub/2dshapedetection-major/TEST'
output_dir = '/home/ashwathama/Documents/GitHub/2dshapedetection-major/CONTOUR_IMAGES'

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Function to remove text heuristically from the image
def remove_text(image):
    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply threshold to get binary image
    _, binary = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)

    # Find contours
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        
        # Heuristic to detect text regions (adjust based on your images)
        aspect_ratio = w / float(h)
        if 0.2 < aspect_ratio < 5 and 10 < w < 100 and 10 < h < 50:
            cv2.rectangle(image, (x, y), (x + w, y + h), (255, 255, 255), -1)

    return image

# Function to process and save images with contours
def process_image(image_path, output_path):
    # Read the image
    image = cv2.imread(image_path)

    # Remove text from the image
    image_no_text = remove_text(image)

    # Convert to grayscale
    gray = cv2.cvtColor(image_no_text, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian Blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Binarization using Otsu's thresholding
    _, binary = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Edge detection using Canny
    edges = cv2.Canny(binary, 50, 150)

    # Finding contours
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours on the original image
    contour_image = image.copy()
    cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)

    # Save the result
    cv2.imwrite(output_path, contour_image)

# Process each image in the input directory
for filename in os.listdir(input_dir):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        input_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, f'contour_{filename}')
        process_image(input_path, output_path)

print(f'Contour images saved in {output_dir}')

```
